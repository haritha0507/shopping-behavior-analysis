## Data visualization

## Univariate

```python
df["Category"].value_counts()
```

```python
df["Category"].value_counts().plot(kind="bar")
plt.title("Most Popular Product Categories")
plt.xlabel("Category")
plt.ylabel("Number of Purchases")
plt.show()
```

```python
# Clothing is the most purchased product category, followed by accessories, while outerwear has the lowest number of purchases.
```

## Bivariate

```python
plt.figure(figsize=(8,4))
sns.countplot(x="Season",data=df, hue="Category")
plt.title("Category by Season")
plt.show()
```

```python
# Across all seasons, clothing remains the most popular category, with accessories showing consistent demand in every season.
```

## multivariate

```python
sns.barplot(x='Season', y='Purchase Amount (USD)', hue="Gender", data=df)
plt.title("Season vs Purchase Amount (USD) by Gender")
plt.show()
```

```python
# The graph shows purchase amounts of males and females in different seasons.Spending changes slightly, with males higher in winter and females higher in summer.
```
